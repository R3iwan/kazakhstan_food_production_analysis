```python
transform_data = []
for city in data['2015']:
    row = {
        'City Name': city,
        '2015': data['2015'][city],
        '2016': data['2016'].get(city, None)
    }
    transform_data.append(row)
    
dataframe = pd.DataFrame(transform_data)
dataframe.to_csv('data_cities.csv', index=False)
```

```python
cities_data = pd.read_csv('data_cities.csv')
```

```python
cities_data
```

```python
plt.figure(figsize=(20,8))
bar_width = 0.4

index = range(len(cities_data['City Name']))

bar1 = plt.bar(index, cities_data['2015'], bar_width, label='2015')
bar2 = plt.bar([i + bar_width for i in index], cities_data['2016'], bar_width, label='2016')

plt.xlabel('City')
plt.ylabel('Prices (KZT) ')
plt.title('Price Comparsion by City from 2015 to 2016')
plt.xticks([i + bar_width / 2 for i in index], cities_data['City Name'], rotation=45)
plt.legend
plt.savefig('price_comparsion_of_rice_from_2015_to_2016.png')

plt.tight_layout()
plt.show()
```
