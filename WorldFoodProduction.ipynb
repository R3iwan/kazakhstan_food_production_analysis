```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('world food production.csv')
```

```python
dataframe.rename(columns={'Rice  Production ( tonnes)': 'RiceProduction'}, inplace=True)
```

```python
kazakhstan_data = dataframe[dataframe['Entity'] == 'Kazakhstan']
```

```python
print(kazakhstan_data.columns)
```

```python
plt.figure(figsize=(30,30))
plt.plot(kazakhstan_data['Year'], kazakhstan_data['RiceProduction'])
plt.xlabel('Year')
plt.ylabel('Amount of Rice Production in tonnes')
plt.title('Graph of Rice Production in Kazakhstan through the years')
plt.grid(True)
plt.savefig('graph_of_rice_production_in_kazakhstan.png')

plt.show
```
